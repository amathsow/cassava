```python
import pandas as pd
import numpy as np
from PIL import Image
import os
from matplotlib import pyplot as plt
```

```python
path = 'extraimages/'
```

```python
def df_images(path):
    im_list = []
    filenames = os.listdir(path)
    for filename in filenames:
        try:
            im = Image.open(path + filename)
            im_list.append([im.filename, im.width, im.height])
        except:
            pass
    df = pd.DataFrame(im_list, columns=['filename','width', 'height'])
    return df
```

```python
df = df_images(path)
df.head()
```
